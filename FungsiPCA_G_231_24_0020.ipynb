```python
def mat_mul(A, B):
    rows_A = len(A)
    cols_A = len(A[0])
    rows_B = len(B)
    cols_B = len(B[0])
    if cols_A != rows_B:
        raise ValueError("Dimensi matriks tidak sesuai.")
    C = [[0 for _ in range(cols_B)] for _ in range(rows_A)]
    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                C[i][j] += A[i][k] * B[k][j]
    return C

def mat_transpose(A):
    """Melakukan transposisi matriks."""
    rows = len(A)
    cols = len(A[0])
    T = [[0 for _ in range(rows)] for _ in range(cols)]
    for i in range(rows):
        for j in range(cols):
            T[j][i] = A[i][j]
    return T

def calculate_column_means(data):
    num_rows = len(data)
    num_cols = len(data[0])
    means = [0.0] * num_cols
    for j in range(num_cols):
        col_sum = 0.0
        for i in range(num_rows):
            col_sum += data[i][j]
        means[j] = col_sum / num_rows
    return means

def center_data(data, means):
    num_rows = len(data)
    num_cols = len(data[0])
    centered_data = [[0.0 for _ in range(num_cols)] for _ in range(num_rows)]
    for i in range(num_rows):
        for j in range(num_cols):
            centered_data[i][j] = data[i][j] - means[j]
    return centered_data

def power_iteration(A, num_simulations=100):
    n = len(A)
    b_k = [1.0 for _ in range(n)]
    for _ in range(num_simulations):
        b_k1 = [0.0 for _ in range(n)]
        for i in range(n):
            for j in range(n):
                b_k1[i] += A[i][j] * b_k[j]
        norm_b_k1 = sum(x**2 for x in b_k1)**0.5
        if norm_b_k1 == 0: break
        b_k = [x / norm_b_k1 for x in b_k1]
    Ab = [0.0 for _ in range(n)]
    for i in range(n):
        for j in range(n):
            Ab[i] += A[i][j] * b_k[j]
    eigenvalue = sum(b_k[i] * Ab[i] for i in range(n))
    return eigenvalue, b_k

def FungsiPCA_NIM(data, n_components=1):
    num_samples = len(data)
    num_features = len(data[0])

    # 1. Mean Centering
    means = calculate_column_means(data)
    X_centered = center_data(data, means)

    # 2. Covariance Matrix
    X_T = mat_transpose(X_centered)
    cov_matrix_raw = mat_mul(X_T, X_centered)
    cov_matrix = [[val / (num_samples - 1) for val in row] for row in cov_matrix_raw]

    # 3. Eigen Decomposition (Power Iteration untuk komponen utama)
    eig_val, eig_vec = power_iteration(cov_matrix)

    # 4. Projection
    projected_data = []
    for row in X_centered:
        dot_product = sum(row[i] * eig_vec[i] for i in range(num_features))
        projected_data.append([dot_product])

    return eig_val, eig_vec, projected_data

# Contoh penggunaan:
data_tes = [[2.5, 2.4], [0.5, 0.7], [2.2, 2.9], [1.9, 2.2], [3.1, 3.0]]
eig_val, eig_vec, hasil = FungsiPCA_NIM(data_tes)
print("Dominant Eigenvalue:", eig_val)
print("Principal Component (Eigenvector):", eig_vec)
print("Projected Data:", hasil)
```
