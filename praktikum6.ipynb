```python
# TUGAS PRAKTIKUM: IMPLEMENTASI EUCLIDEAN DISTANCE
import math
import numpy as np

# --- 1. Cara Pertama: Menggunakan Python Murni (Looping & Math library) ---
def euclidean_distance_pure(x, y):
    """
    Menghitung Euclidean Distance antara dua list/vektor menggunakan Python murni.
    """
    if len(x) != len(y):
        raise ValueError("Dimensi kedua vektor harus sama!")

    jumlah_kuadrat = 0
    for i in range(len(x)):
        jumlah_kuadrat += (x[i] - y[i]) ** 2

    return math.sqrt(jumlah_kuadrat)


# --- 2. Cara Kedua: Menggunakan Library NumPy (Lebih Efisien & Standar Data Science) ---
def euclidean_distance_numpy(x, y):
    """
    Menghitung Euclidean Distance menggunakan operasi vektor NumPy.
    """
    vektor_x = np.array(x)
    vektor_y = np.array(y)

    # np.linalg.norm langsung menghitung akar dari jumlah kuadrat selisihnya
    return np.linalg.norm(vektor_x - vektor_y)


# CONTOH PENGGUNAAN (TEST CASE)

# Kita coba dengan titik 3 dimensi (x1, x2, x3) dan (y1, y2, y3)
titik_A = [3, 4, 5]
titik_B = [1, 2, 1]

print("--- Hasil Perhitungan Euclidean Distance ---")

# Eksekusi Cara 1
hasil_pure = euclidean_distance_pure(titik_A, titik_B)
print(f"Hasil (Python Murni) : {hasil_pure:.4f}")

# Eksekusi Cara 2
hasil_numpy = euclidean_distance_numpy(titik_A, titik_B)
print(f"Hasil (NumPy)        : {hasil_numpy:.4f}")
```
