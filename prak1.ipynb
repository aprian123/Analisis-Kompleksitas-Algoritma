```python
# Fungsi untuk menghitung jumlah elemen array
def list_sum(arr, n):
    total_sum = 0                       # Inisialisasi awal jumlah = 0 [cite: 216]
    for i in range(n):                  # Melakukan perulangan sebanyak n elemen [cite: 218]
        total_sum = total_sum + arr[i]  # Menambahkan elemen ke-i ke total jumlah [cite: 220]
    return total_sum                    # Mengembalikan hasil akhir [cite: 222]

# Driver Program (Contoh Penggunaan)
arr = [5, 6, 1, 2]                      # Input array [cite: 224]
n = len(arr)                            # Mengambil panjang array [cite: 225]

print("--- LATIHAN 1: SUM ARRAY ---")
print("Input Array :", arr)
print("Output Sum  :", list_sum(arr, n)) # Hasilnya adalah 14 [cite: 226]
```

```python
# Driver Program
n = 3 # Jumlah baris [cite: 239]
m = 3 # Jumlah kolom [cite: 240]

# Input matriks 2D [cite: 241]
matrix = [
    [3, 2, 7],
    [2, 6, 8],
    [5, 1, 9]
]

total_matrix_sum = 0 # Inisialisasi awal jumlah = 0 [cite: 242]

# Loop luar untuk iterasi baris (n kali) [cite: 245, 257]
for i in range(n):
    # Loop dalam untuk iterasi kolom (m kali) [cite: 246, 257]
    for j in range(m):
        total_matrix_sum += matrix[i][j] # Menjumlahkan elemen baris ke-i dan kolom ke-j [cite: 255]

print("\n--- LATIHAN 2: SUM MATRIX ---")
print("Matrix:")
for row in matrix:
    print(row)
print("Output Total Semua Elemen Matriks:", total_matrix_sum) # Hasilnya adalah 43 [cite: 254]
```

```python
# Fungsi Binary Search
def binary_search(sorted_list, target):
    low = 0                           # Batas bawah awal
    high = len(sorted_list) - 1       # Batas atas awal

    while low <= high:                # Perulangan selama wilayah pencarian masih valid
        mid = (low + high) // 2       # Menemukan indeks tengah
        guess = sorted_list[mid]      # Mengambil nilai di posisi tengah

        if guess == target:
            return mid                # Target ditemukan, kembalikan posisi indeksnya
        if guess > target:
            high = mid - 1            # Jika tebakan terlalu besar, geser batas atas ke kiri
        else:
            low = mid + 1             # Jika tebakan terlalu kecil, geser batas bawah ke kanan

    return -1                         # Mengembalikan -1 jika target tidak ditemukan

# Driver Program (Contoh Penggunaan)
my_list = [1, 3, 5, 7, 9, 11, 13, 15] # List harus sudah terurut
target_val = 7                        # Angka yang ingin dicari

print("\n--- LATIHAN 3: BINARY SEARCH ---")
result = binary_search(my_list, target_val) # Memanggil fungsi pencarian

print(f"Input: {my_list}, Target: {target_val}")
if result != -1:
    print(f"Output: Elemen ditemukan dalam indeks {result}") # Hasilnya indeks 3
else:
    print("Output: Elemen tidak ditemukan")
```
