```python
# =====================================================================
# TUGAS 1: IMPLEMENTASI RED-BLACK TREE
# =====================================================================

class Node:
    def __init__(self, data):
        self.data = data
        self.parent = None
        self.left = None
        self.right = None
        self.color = 1  # 1 mewakili RED, 0 mewakili BLACK

class RedBlackTree:
    def __init__(self):
        # T.NIL digunakan sebagai node eksternal (daun kosong berwujud BLACK)
        self.NIL = Node(0)
        self.NIL.color = 0
        self.root = self.NIL

    # --- Fungsi Rotasi Kiri ---
    def left_rotate(self, x):
        y = x.right
        x.right = y.left
        if y.left != self.NIL:
            y.left.parent = x
        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y
        y.left = x
        x.parent = y

    # --- Fungsi Rotasi Kanan ---
    def right_rotate(self, x):
        y = x.left
        x.left = y.right
        if y.right != self.NIL:
            y.right.parent = x
        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.right:
            x.parent.right = y
        else:
            x.parent.left = y
        y.right = x
        x.parent = y

    # --- Perbaikan Pohon Setelah Insert (Fix-Up) ---
    def fix_insert(self, k):
        while k.parent.color == 1:
            if k.parent == k.parent.parent.right:
                u = k.parent.parent.left  # paman k
                if u.color == 1: # Kasus 1: Paman berwarna Merah
                    u.color = 0
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    k = k.parent.parent
                else:
                    if k == k.parent.left: # Kasus 2: Paman Hitam & k berbentuk segitiga
                        k = k.parent
                        self.right_rotate(k)
                    # Kasus 3: Paman Hitam & k berbentuk garis lurus
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    self.left_rotate(k.parent.parent)
            else:
                u = k.parent.parent.right  # paman k
                if u.color == 1: # Kasus 1 alternatif
                    u.color = 0
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    k = k.parent.parent
                else:
                    if k == k.parent.right: # Kasus 2 alternatif
                        k = k.parent
                        self.left_rotate(k)
                    # Kasus 3 alternatif
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    self.right_rotate(k.parent.parent)
            if k == self.root:
                break
        self.root.color = 0 # Akar harus selalu Hitam

    # --- Fungsi Memasukkan Node Baru (Insert) ---
    def insert(self, key):
        node = Node(key)
        node.parent = None
        node.data = key
        node.left = self.NIL
        node.right = self.NIL
        node.color = 1  # Node baru selalu berwarna MERAH

        y = None
        x = self.root

        while x != self.NIL:
            y = x
            if node.data < x.data:
                x = x.left
            else:
                x = x.right

        node.parent = y
        if y == None:
            self.root = node
        elif node.data < y.data:
            y.left = node
        else:
            y.right = node

        if node.parent == None:
            node.color = 0
            return

        if node.parent.parent == None:
            return

        self.fix_insert(node)

    # --- Fungsi Pembantu Cetak Struktur Pohon ---
    def print_tree_helper(self, node, indent, last):
        if node != self.NIL:
            print(indent, end="")
            if last:
                print("R----", end="")
                indent += "     "
            else:
                print("L----", end="")
                indent += "|    "

            color_str = "RED" if node.color == 1 else "BLACK"
            print(f"{node.data}({color_str})")
            self.print_tree_helper(node.left, indent, False)
            self.print_tree_helper(node.right, indent, True)

    def print_tree(self):
        self.print_tree_helper(self.root, "", True)


# --- DRIVER PROGRAM UNTUK UJI COBA RBT ---
rbt = RedBlackTree()
data_input = [10, 20, 30, 15, 25]

print("\n--- PROSES INSERT RED-BLACK TREE ---")
for val in data_input:
    print(f"\nSetelah memasukkan angka {val}:")
    rbt.insert(val)
    rbt.print_tree()
```

1. Kasus Terburuk (Worst Case): $O(\log n)$   
  Penjelasan: Berbeda dengan Binary Search Tree biasa yang bisa menjadi timpang sebelah (skewed), aturan rotasi dan pewarnaan pada Red-Black Tree menjamin tinggi pohon maksimal selalu berkisar $2 \log(n + 1)$. Pencarian atau penyisipan elemen terlama pun tetap berjalan dalam waktu logaritmik.

2. Kasus Rata-rata (Average Case): $O(\log n)$
Penjelasan: Secara acak, distribusi data akan membentuk pohon biner seimbang yang ideal, membuat operasi pencarian, penambahan, dan penghapusan data membutuhkan waktu proporsional terhadap tinggi pohonnya.
