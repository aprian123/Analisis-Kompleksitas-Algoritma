```python
# =====================================================================
# 1. FIBONACCI REKURSIF BIASA (TANPA MEMOIZATION)
# =====================================================================
def fib_recursive(n):
    if n <= 1: # cite: 14
        return n # cite: 16
    return fib_recursive(n-1) + fib_recursive(n-2) # cite: 18

# Driver Program
n_rec = 10 # cite: 21
print("=== Fibonacci Rekursif Biasa ===")
print("Fibonacci:", fib_recursive(n_rec)) # cite: 22
print("-" * 40)


# =====================================================================
# 2. FIBONACCI DENGAN MEMOIZATION (TOP-DOWN)
# =====================================================================
memo = {} # cite: 32

def fib_memo(n):
    if n in memo: # cite: 35
        return memo[n] # cite: 38
    if n <= 1: # cite: 42
        return n # cite: 44

    memo[n] = fib_memo(n-1) + fib_memo(n-2) # cite: 49
    return memo[n] # cite: 49

# Driver Program
n_memo = 50 # cite: 51
print("=== Fibonacci Memoization ===")
print("Fibonacci:", fib_memo(n_memo)) # cite: 52
print("-" * 40)

# =====================================================================
# 3. FIBONACCI DYNAMIC PROGRAMMING (BOTTOM-UP)
# =====================================================================
def fib_dp(n):
    if n <= 1: # cite: 65
        return n # cite: 69

    dp = [0] * (n + 1) # cite: 72
    dp[0] = 0 # cite: 75
    dp[1] = 1 # cite: 77

    for i in range(2, n + 1): # cite: 84
        dp[i] = dp[i-1] + dp[i-2] # cite: 90

    return dp[n] # cite: 85

# Driver Program
n_dp = 50 # cite: 87
print("=== Fibonacci Dynamic Programming ===")
print("Fibonacci DP:", fib_dp(n_dp)) # cite: 88
print("-" * 40)


# =====================================================================
# 4. FIBONACCI DP - SPACE OPTIMIZED
# =====================================================================
def fib_optimized(n):
    if n <= 1: # cite: 103
        return n # cite: 105

    a, b = 0, 1 # cite: 108
    for _ in range(2, n + 1): # cite: 111, 112
        a, b = b, a + b # cite: 114
    return b # cite: 117

# Driver Program
n_opt = 50 # cite: 120
print("=== Fibonacci DP Space Optimized ===")
print("Optimized Fibonacci:", fib_optimized(n_opt)) # cite: 121
print("-" * 40)

# =====================================================================
# 5. GREEDY COIN CHANGE
# =====================================================================
def coin_change_greedy(coins, amount):
    coins.sort(reverse=True) # Urutkan koin dari yang terbesar (cite: 136)
    result = [] # cite: 146

    for coin in coins: # cite: 147
        while amount >= coin: # cite: 148
            result.append(coin) # cite: 149
            amount -= coin # Mengurangi jumlah sisa uang (perbaikan sintaks OCR dari baris 150)

    return result # cite: 153

# Driver Program
coins_greedy = [100, 50, 25, 10, 5, 1] # cite: 156
amount_greedy = 289 # cite: 157
hasil_greedy = coin_change_greedy(coins_greedy, amount_greedy) # cite: 161

print("=== Greedy Coin Change ===")
print("Koin digunakan:", hasil_greedy) # cite: 162
print("Jumlah koin:", len(hasil_greedy)) # cite: 163
print("-" * 40)

# =====================================================================
# 6. DP COIN CHANGE (MENGHITUNG JUMLAH KOMBINASI)
# =====================================================================
def count_ways(coins, amount):
    dp = [0] * (amount + 1) # cite: 195
    dp[0] = 1 # Base case: ada 1 cara untuk membuat nominal 0 (cite: 196)

    for coin in coins: # cite: 197
        for i in range(coin, amount + 1): # cite: 198
            dp[i] += dp[i - coin] # cite: 199

    return dp[amount] # cite: 200

# Driver Program
coins_dp = [1, 2, 5] # cite: 202
amount_dp = 5 # cite: 203
print("=== DP Coin Change (Combination Ways) ===")
print("Jumlah kombinasi:", count_ways(coins_dp, amount_dp)) # cite: 205
print("=" * 40)
```
