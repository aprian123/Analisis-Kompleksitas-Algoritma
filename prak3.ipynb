```python
# =====================================================================
# TUGAS: IMPLEMENTASI JARK TERPENDEK (DIJKSTRA ALGORITHM)
# =====================================================================
import sys

def dijkstra(graph, start_node):
    # 1. Inisialisasi jarak ke semua simpul sebagai tak hingga (infinity)
    #    dan jarak ke simpul asal (start_node) sebagai 0
    shortest_distances = {node: sys.maxsize for node in graph}
    shortest_distances[start_node] = 0

    # 2. Himpunan untuk melacak simpul yang jarak terpendeknya sudah dipastikan
    visited_nodes = set()

    # 3. Struktur data untuk melacak jalur asal node (opsional, untuk rekonstruksi rute)
    predecessors = {node: None for node in graph}

    while len(visited_nodes) < len(graph):
        # Cari simpul dengan jarak terkecil dari simpul yang belum dikunjungi
        current_node = None
        min_distance = sys.maxsize

        for node in graph:
            if node not in visited_nodes and shortest_distances[node] < min_distance:
                min_distance = shortest_distances[node]
                current_node = node

        # Jika tidak ada simpul yang bisa dijangkau lagi, keluar dari loop
        if current_node is None:
            break

        # Tandai simpul saat ini sebagai simpul yang sudah dikunjungi
        visited_nodes.add(current_node)

        # Perbarui nilai jarak untuk semua simpul tetangga dari simpul saat ini
        for neighbor, weight in graph[current_node].items():
            # Hitung bobot/jarak baru jika melewati current_node
            new_distance = shortest_distances[current_node] + weight

            # Jika jalur baru ini lebih pendek, perbarui datanya
            if new_distance < shortest_distances[neighbor]:
                shortest_distances[neighbor] = new_distance
                predecessors[neighbor] = current_node

    return shortest_distances, predecessors

# --- DRIVER PROGRAM UNTUK UJI COBA TUGAS ---
#Representasi graf menggunakan Adjancency List (Daftar Kedekatan)
# Contoh Graf: A memiliki tetangga B (bobot 4) dan C (bobot 2), dst.
graf_sample = {
    'A': {'B': 4, 'C': 2},
    'B': {'C': 1, 'D': 5},
    'C': {'B': 1, 'D': 8, 'E': 10},
    'D': {'E': 2},
    'E': {}
}

simpul_awal = 'A'
jarak_terpendek, rute_asal = dijkstra(graf_sample, simpul_awal)

print("\n--- TUGAS: HASIL PENCARIAN JARAK TERPENDEK (DIJKSTRA) ---")
print(f"Jarak terpendek dari simpul asal '{simpul_awal}':")
for simpul, jarak in jarak_terpendek.items():
    nilai_jarak = jarak if jarak != sys.maxsize else "Tidak terjangkau"
    print(f"Ke Simpul {simpul} = Jarak: {nilai_jarak}")
```

1. Analisis Kompleksitas Waktu (Time Complexity):
Menggunakan
array/objek biasa untuk mencari simpul dengan bobot minimum (seperti contoh di atas), kompleksitas waktunya adalah $O(V^2)$, di mana $V$ mewakili jumlah simpul (Vertices) di dalam graf. Hal ini dikarenakan program melakukan pencarian linear untuk simpul terdekat pada setiap tahap iterasi.Catatan tambahan: Kompleksitas ini dapat dioptimasi menjadi $O((V + E) \log V)$ jika menggunakan struktur data Min-Priority Queue (seperti modul heapq bawaan Python), di mana $E$ mewakili jumlah garis hubung (Edges).
2. Analisis Kompleksitas Ruang (Space Complexity):
Kompleksitas ruangnya adalah $O(V + E)$ untuk menyimpan representasi struktur matriks/daftar kedekatan graf serta variabel penampung riwayat jarak simpul yang sedang diperiksa di dalam memori komputer.
