```python
# =====================================================================
# ALGORITMA 1: LINEAR SEARCH
# =====================================================================
def linear_search(arr, target):
    for i in range(len(arr)):        # Iterasi dari indeks 0 hingga akhir
        if arr[i] == target:         # Jika elemen ditemukan
            return i                 # Kembalikan indeksnya
    return -1                        # Kembalikan -1 jika tidak ditemukan

# Driver Program / Contoh Penggunaan
data_linear = [10, 20, 30, 40, 50]
target_linear = 30

print("--- 1. LINEAR SEARCH ---")
print("Data   :", data_linear)
print("Target :", target_linear)
print("Output Indeks:", linear_search(data_linear, target_linear)) # Output: 2
```

```python
# =====================================================================
# ALGORITMA 2: BINARY SEARCH (ITERATIF)
# =====================================================================
def binary_search(arr, target):
    left, right = 0, len(arr) - 1    # Inisialisasi batas kiri dan kanan

    while left <= right:
        mid = (left + right) // 2    # Menghitung indeks tengah

        if arr[mid] == target:       # Target ditemukan di tengah
            return mid
        elif arr[mid] < target:      # Jika target lebih besar, geser batas kiri
            left = mid + 1
        else:                        # Jika target lebih kecil, geser batas kanan
            right = mid - 1

    return -1                        # Target tidak ditemukan

# Driver Program (Data harus terurut)
data_binary = [10, 20, 30, 40, 50]
target_binary = 40

print("\n--- 2. BINARY SEARCH (ITERATIF) ---")
print("Data   :", data_binary)
print("Target :", target_binary)
print("Output Indeks:", binary_search(data_binary, target_binary)) # Output: 3
```

```python
# =====================================================================
# ALGORITMA 3: BINARY SEARCH (REKURSIF)
# =====================================================================
def binary_search_recursive(arr, left, right, target):
    if left > right:                 # Base case: wilayah pencarian habis
        return -1

    mid = (left + right) // 2        # Menghitung indeks tengah

    if arr[mid] == target:           # Target ditemukan
        return mid
    elif arr[mid] < target:          # Cari ke paruh kanan graf/array
        return binary_search_recursive(arr, mid + 1, right, target)
    else:                            # Cari ke paruh kiri array
        return binary_search_recursive(arr, left, mid - 1, target)

# Driver Program
data_recursive = [10, 20, 30, 40, 50]
target_recursive = 20

print("\n--- 3. BINARY SEARCH (REKURSIF) ---")
print("Data   :", data_recursive)
print("Target :", target_recursive)
# Parameter left dimulai dari 0 dan right dari len(data)-1
print("Output Indeks:", binary_search_recursive(data_recursive, 0, len(data_recursive) - 1, target_recursive)) # Output: 1
```

```python
# =====================================================================
# ALGORITMA 4: JUMP SEARCH
# =====================================================================
import math

def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))         # Menentukan ukuran lompatan (blok)
    prev = 0                         # Menyimpan posisi awal blok

    # Melompat per blok selama elemen di akhir blok masih lebih kecil dari target
    while arr[min(step, n) - 1] < target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1                # Target tidak ada di dalam jangkauan array

    # Melakukan Linear Search di dalam blok yang sudah ditentukan
    for i in range(prev, min(step, n)):
        if arr[i] == target:
            return i                 # Target ditemukan

    return -1                        # Target tidak ditemukan

# Driver Program
data_jump = [10, 20, 30, 40, 50, 60, 70]
target_jump = 60

print("\n--- 4. JUMP SEARCH ---")
print("Data   :", data_jump)
print("Target :", target_jump)
print("Output Indeks:", jump_search(data_jump, target_jump)) # Output: 5
```

```python
# =====================================================================
# ALGORITMA 5: INTERPOLATION SEARCH
# =====================================================================
def interpolation_search(arr, target):
    low = 0
    high = len(arr) - 1

    # Syarat perulangan: target harus berada di dalam rentang nilai low dan high
    while low <= high and target >= arr[low] and target <= arr[high]:
        if low == high:
            if arr[low] == target:
                return low
            return -1

        # Rumus Estimasi Posisi Kunci (Probe Position Formula)
        pos = low + int(((float(high - low) / (arr[high] - arr[low])) * (target - arr[low])))

        if arr[pos] == target:       # Target ditemukan
            return pos
        elif arr[pos] < target:      # Jika estimasi terlalu kecil, geser batas bawah
            low = pos + 1
        else:                        # Jika estimasi terlalu besar, geser batas atas
            high = pos - 1

    return -1

# Driver Program
data_interpolation = [10, 20, 30, 40, 50, 60, 70]
target_interpolation = 50

print("\n--- 5. INTERPOLATION SEARCH ---")
print("Data   :", data_interpolation)
print("Target :", target_interpolation)
print("Output Indeks:", interpolation_search(data_interpolation, target_interpolation)) # Output: 4
```

1. Linear Search <br> Best Case: $O(1)$ (Data ditemukan di indeks pertama).
Worst Case: $O(n)$ (Data berada di paling ujung atau tidak ada sama sekali).
2. Binary Search (Iteratif & Rekursif)<br>Best Case: $O(1)$ (Data langsung ditemukan tepat di tengah array).<br>Worst Case: $O(\log n)$ (Wilayah pencarian harus terus dibagi dua sampai sisa 1 elemen).
3. Jump Search<br>Best Case: $O(1)$ (Data berada di indeks paling awal).<br>Worst Case: $O(\sqrt{n})$ (Pencarian harus melompati seluruh blok dan melakukan pencarian linear penuh di blok terakhir).
4. Interpolation Search<br>Best Case (Average Case): $O(\log(\log n))$ (Data terdistribusi seragam sehingga pencarian sangat cepat).<br>Worst Case: $O(n)$ (Data terdistribusi tidak merata/eksponensial, sehingga rumusnya meleset terus dan berubah menjadi linear search biasa).
